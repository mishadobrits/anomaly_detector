```python
dfs = [dfar1, dfar2, dfap1, dfap2]
```

```python
dfar2 = dfar2.reindex(sorted(dfar2.columns), axis=1)
```

```python
drop = list(set(dfap1.columns).symmetric_difference(dfap2.columns))
drop.extend(['AOM', 'VAE', 'MAD', 'LSCP'])
```

```python
dfar = pd.concat([dfar1, dfar2], ignore_index=True)
dfar.drop(columns = drop, inplace = True)
dfar = dfar.rename({'Unnamed: 0': 'Dataset'}, axis=1)
dfar = dfar.drop_duplicates()
#dfar = dfar['Dataset'].map({'2_annthyroid':'02_annthyroid', '6_cardio': '06_cardio'})
dfar['Dataset'][5] = '02_annthyroid'
dfar['Dataset'][0] = '06_cardio'
dfar = dfar.sort_values(by = ['Dataset'], ignore_index=True)
dfar = dfar.set_index('Dataset')
dfar
```

```python
dfar.isna().sum().sum()
```

```python
dfap = pd.concat([dfap1, dfap2], ignore_index=True)
dfap.drop(columns = drop, inplace = True)
dfap = dfap.rename({'Unnamed: 0': 'Dataset'}, axis=1)
dfap = dfap.drop_duplicates()
#dfar = dfar['Dataset'].map({'2_annthyroid':'02_annthyroid', '6_cardio': '06_cardio'})
dfap['Dataset'][5] = '02_annthyroid'
dfap['Dataset'][0] = '06_cardio'
dfap = dfap.sort_values(by = ['Dataset'], ignore_index=True)
dfap = dfap.set_index('Dataset')
dfap
```

```python
def highlight_max_09(s, props=''):
    return np.where(10 * s >= 9 * np.nanmax(s.values), props, '')
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
star = dfar.style.apply(highlight_max_09, props='background-color:#00d000', axis=1).apply(highlight_max, props='font-weight:bold', axis=1)
star
```

```python
stap = dfap.style.apply(highlight_max_09, props='background-color:#00d000', axis=1).apply(highlight_max, props='font-weight:bold', axis=1)
stap
```

```python
dfar.to_csv('AUCROC.csv', index=False)
dfap.to_csv('AUCPR.csv', index=False)
```
